```python
## Import Libraries
import pandas as pd
## read csv file and rename column
RawData=pd.read_csv("train.csv",
                    header=0,
                   names=["Passenger_Id","If_Survived","Ticket_Class","Name","Sex","Age","Num_Aboard_Sibling_Spouses","Num_Aboard_Parents_Children","Ticket_Num","Fare","Cabin","Embarked_Loc"],
                   index_col=["Passenger_Id"])
RawData.info()
```

### 1-3 String Feature Engineering

## 2- Statistic Analysis between Cabin features and Survival Rate

```python
# Which Cabin_Letter give the highest chance to survive? 
CabinLetterGroupDF=df1_Splited.groupby(['Cabin_Letter'])['If_Survived'].aggregate(['sum','count','mean'])
CabinLetterGroupDF.columns=['SurvivalCount','Total','SurvivalRatio']
CabinLetterGroupDF['SurvivalRatio'].dtype
CabinLetterGroupDF['SurvivalRatio'] = CabinLetterGroupDF['SurvivalRatio'].astype(float).map("{:.2%}".format)
CabinLetterGroupDF.sort_values('SurvivalRatio',ascending=False)
```

## 3- Data Cleaning

### 3-1 Drop or Fill the Missing Vaue

#### Drop Entire Cabin Column and Embarked Location Missing Value Row

```python
#Drop the Cabin column 
df2=RawData.drop(['Cabin'], axis=1)
#Drop the record where Embarked_Loc is empty (only 2/891 lines)
df2=df2.dropna(axis=0, how='any', thresh=None, subset=["Embarked_Loc"], inplace=False)
df2.info()
```

#### Fill Age N/A by Gender

```python
df2["Age"].describe()
```

```python
# Get Age median by Sex
AgebyGender=df2.groupby(['Sex'])['Age'].aggregate(['median','mean'])
AgebyGender
```

```python
# Fill the missing value using the Age median by Sex
df2_female=df2[df2['Sex']=="female"]
df2_female["Age"].fillna(df2_female["Age"].median(),inplace=True) 
df2_female.info()
```

```python
df2_male=df2[df2['Sex']=="male"]
df2_male["Age"].fillna(df2_male["Age"].median(),inplace=True) 
df2_male.info()
```

```python
#Append df2_female and df2_male
df2_filled=pd.concat([df2_female,df2_male],axis=0)
df2_filled.info()
```

### 3-3 Feature Engineering

```python
#Add Age Group
Age=pd.DataFrame(data=df2_filled["Age"])
#Age = pd.cut(Age["Age"], 10, right=True, labels=None)
bins = [0,8,16,24,32,40,48,56,64,72,80]
bin_labels = ['[0,8)','[8,16)','[16,24)','[24,32)', '[32,40)', '[40,48)', '[48,56)','[56,64)','[64,72)','>=72']
Age["Age_Group"] = pd.cut(Age["Age"], bins, right=False, labels = bin_labels) 
Age=Age.drop(['Age'], axis=1)
df2_AgeGroup=df2_filled.join(Age,sort=True)
df2_AgeGroup.info()
```

### Multi Factor Analysis

```python
# Cross Tab Survival Rate Analysis: Embarked_Loc & Ticket_Class
Loc_Class_Analysis = pd.crosstab(df2_AgeGroup['Embarked_Loc'], df2_AgeGroup['Ticket_Class'], df2_AgeGroup['If_Survived'], aggfunc='mean').round(1)
Loc_Class_Analysis
#---Most people aboard from Cherbourg are in Upper Class 
```

```python
#Why female has higher survival rate? 
# Cross Tab Analysis: Sex among age group distribution
Sex_AG_Distri = pd.crosstab(df2_AgeGroup['Age_Group'], df2_AgeGroup['Sex'], df2_AgeGroup['If_Survived'], aggfunc='count').round(1)
#Sex_AG_Distri.plot(kind='bar', figsize=(10,5))
Sex_AG_Distri=Sex_AG_Distri.apply(lambda x: x/x.sum())
Sex_AG_Distri['female'] = Sex_AG_Distri['female'].astype(float).map("{:.2%}".format)
Sex_AG_Distri['male'] = Sex_AG_Distri['male'].astype(float).map("{:.2%}".format)
Sex_AG_Distri
#---female are all under 64 versus oldest male is 80
#---more female fall under age basket under 16, where get help from parents
```

```python
# Cross Tab Survival Rate Analysis: Age Group & Sex 
Sex_AG_Analysis = pd.crosstab(df2_AgeGroup['Age_Group'], df2_AgeGroup['Sex'], df2_AgeGroup['If_Survived'], aggfunc='mean').round(1)
Sex_AG_Analysis
#---female across are age groups are get higher survival rate. Will they get more help from others? genes?
```
